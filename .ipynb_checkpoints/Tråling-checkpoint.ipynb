## Tråling i et definert korpus

Finner bøker som inneholder varianter av ord, og sammenligner. 


Starter med å definere et korpus

```python
korpus = nb.book_urn(words='arbeid', ddk="3%", period=(1960, 1970), limit=1000)
```

```python
len(korpus)
```

Dernest kommer settet det skal tråles med.

```python
wb = {'Kvinne': [
    'kvinne', 'kvinner', 'kvinnene'
],
     'Frihet':['frihet', 'grenser', 'fengsel'],
      
     'Lønn':['godtgjørelse', 'kapital', 'lønn', 'lønna', 'lønnen']}
```

Selve trålingen gjøres med kommandoen `wordbag_eval`

```python
trål = nb.wordbag_eval(wb, korpus)
```

Se på resultatene. De er sortert etter snittverdienm

```python
trål
```

For å begrense kan det lages forskjellige kommandoer. Eksempler

```python
# Se på der hvor verdien er større enn en gitt verdi

trål[(trål['Kvinne'] > 100) & (trål['Frihet'] > 10) & (trål['Lønn'] > 4)]
```

Resultatet kan legges i en variabel


```python
fangst = trål[(trål['Kvinne'] > 100) & (trål['Frihet'] > 10) & (trål['Lønn'] > 4)]
```

Og metadata kan hentes ut

```python
nb.frame(nb.metadata(list(fangst.index)))
```
